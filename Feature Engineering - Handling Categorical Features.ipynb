```python
uci_data
```

#### Finding number of unique categories in each feature

```python
for feature in uci_data[:]:
    print(feature,":",len(uci_data[feature].unique()),"labels")
```

```python
uci_data['Country'].value_counts().to_dict()
```

#### Now we will replace the country category with their frequency value counts. Ex: India:100, Japan:62

```python
country_map = uci_data['Country'].value_counts().to_dict()
```

```python
uci_data['Country'] = uci_data['Country'].map(country_map)
```

```python
uci_data.columns[-1]
```

```python
uci_data
```

#### Advantages

1. Easy to Use
2. Not increasing feature space or dimensions

#### Disadvantages

1. It will provide same weight if frequencies are same

#### Target Guided Ordinal Encoding

What is it and When it will be used?

1. Ordering the variables according to the target
2. Replace the variables by the joint probability of being 1 or 0

```python
df3 = pd.read_csv('titanic.csv',usecols=['Cabin','Survived'])
```

```python
df3
```

```python
df3['Cabin'].fillna("Missing",inplace=True)
```

```python
df3
```

```python
df3['Cabin'] = df3['Cabin'].astype(str).str[0]
```

```python
df3
```

```python
df3['Cabin'].unique()
```

```python
df3.groupby('Cabin')['Survived'].mean()
```

```python
df3.groupby('Cabin')['Survived'].mean().sort_values()
```

```python
df3.groupby('Cabin')['Survived'].mean().sort_values().index
```

```python
ordinal_labels = df3.groupby('Cabin')['Survived'].mean().sort_values().index
```

```python
ordinal_labels
```

```python
ordinal_labels2 = {k:i for i,k in enumerate(ordinal_labels,0)}
```

```python
enumerate(ordinal_labels,0)
```

```python
ordinal_labels2
```

```python
df3['Cabin_ordinal_labels'] = df3['Cabin'].map(ordinal_labels2)
```

```python
df3
```

#### Probability ratio encoding

1. Probability of Survived based on Cabin -- Categorical Feature
2. Probability of not-survived --> 1-pr(Survived)
3. pr(Survived)/pr(Not survived)
4. Dictionary to map cabin with probability
5. Replace with the categorical feature

```python
df4 = pd.read_csv('titanic.csv',usecols=['Cabin','Survived'])
```

```python
df4
```

```python
#### Replacing nan with missing
df4['Cabin'].fillna("Missing",inplace=True)
```

```python
df4
```

```python
df4['Cabin'].unique()
```

```python
df4['Cabin'] = df4['Cabin'].astype(str).str[0]
```

```python
df4
```

```python
df4['Cabin'].unique()
```

```python
prob_df = df4.groupby(['Cabin'])['Survived'].mean()
```

```python
prob_df = pd.DataFrame(prob_df)
```

```python
prob_df
```

```python
prob_df['Died'] = 1-prob_df['Survived']
```

```python
prob_df
```

```python
prob_df['Probability_ratio'] = prob_df['Survived']/prob_df['Died']
```

```python
prob_df
```

```python
prob_df['Probability_ratio'].to_dict()
```

```python
probabilty_encoded = prob_df['Probability_ratio'].to_dict()
```
