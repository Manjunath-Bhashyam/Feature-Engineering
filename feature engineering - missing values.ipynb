#### What are the different types of Missing Data?

##### Type 1. Missing Completely at Random, MCAR:
A variable is missing completely at random (MCAR) if the probability of being missing is the same for all the observations. When data is MCAR, there is absolutely no relationship between the data missing and any other values, observed or missing, within the dataset. In other words, those missing data points are a random subset of the data. There is nothing systematic going on that makes some data more likely to be missing than other.

#### Observation: There is some relationship between Age and Cabin as  the data is recorded after the accident occurred and many of the family members might have dead already and they might have been in those cabin for which the data is missing for it. Hence it is not a category of missing values at random.

#### Observation: Embarked two values are missing and there is no relationship with any of other features, so it is an example of missing completely at random (MCAR)

##### Type 2.  Missing Data Not At Random(MNAR): Systematic missing Values
There is absolutely some relationship between the data missing and any other values, observed or missing, within the dataset.

### Mean/ Median/ Mode Imputation - When should we need to apply
Mean/ Median/ Mode has the assumption that the data are missing completely at random (MCAR). We solve this by replacing the NAN with the most frequent occurance of the variables

```python
df=pd.read_csv('titanic.csv',usecols=['Age','Fare','Survived'])
df.head()
```

```python
## Lets go and see the percentage of missing values
np.round(df.isnull().mean()*100,4)
```

```python
def impute_nan(df,variable,median):
    df[variable+'_median'] = df[variable].fillna(median)
```

```python
median = df['Age'].median()
median
```

```python
impute_nan(df,'Age',median)
```

### Advantages And Disadvantages of Mean/Median Imputation

#### Advantages
1. Easy to implement(Robust to outliers)
2. Faster way to obtain the complete dataset
#### Disadvantages
1. Change or Distortion in the original variance
2. Impacts Correlation

#### 2. Random Sample Imputation

AIM: Random Sample Imputation consists of taking random sample observation from dataset and it picks randomly and we use this observation to fill the NAN values

When should it be used?
It assumes the data is missing completely at random (MCAR)

```python
df1 = pd.read_csv("titanic.csv",usecols=['Survived','Age','Fare'])
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
np.round(df1.isnull().mean()*100,2)
```

```python
#Selecting a random sample to fillna
```

```python
df1['Age'].dropna().sample()
```

```python
df1['Age'].isnull().sum() #No. of NAN Values
```

```python
df1['Age'].dropna().sample(df1['Age'].isnull().sum(),random_state=101) # All 177 Nan Values are filled with random sample observations
```

```python
def impute_nan(df,variable,median):
    df[variable+'_median'] = df[variable].fillna(median)
    df[variable+'_random'] = df[variable]
    ## Random Sample for fill na wih random observations
    random_sample = df['Age'].dropna().sample(df['Age'].isnull().sum(),random_state=101)
    ## Pandas need to have same index in order to merge the dataset with random sample created above
    random_sample.index = df[df[variable].isnull()].index
    df.loc[df[variable].isnull(),variable+'_random'] = random_sample
```

```python
impute_nan(df1,'Age',median)
```

```python
df1
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
df1['Age'].plot(kind='kde', ax=ax)
df1['Age_random'].plot(kind='kde', ax=ax, color='red')
lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
```

#### 4. End of Distribution Imputation

Where to be used? It works well when missing values not at random

```python
df4 = pd.read_csv("titanic.csv",usecols=['Survived','Age','Fare'])
```

```python
df4.head()
```

```python
df4['Age'].hist(bins=50)
```

```python
extreme = df4['Age'].mean()+3*df4['Age'].std()
```

```python
sns.boxplot(data=df4,x=df4['Age'])
```

```python
def impute_nan(df,variable,median,extreme):
    df[variable+'_end_distribution'] = df[variable].fillna(extreme)
    df[variable].fillna(median,inplace=True)
```

```python
impute_nan(df4,'Age',df4['Age'].median(),extreme)
```

```python
df4
```

```python
df4['Age'].hist(bins=50)
```

```python
df4['Age_end_distribution'].hist(bins=50)
```

#### Advantages

1. Easy to Implement
2. Captures the importance of missingness if there is one

#### Disadvantages

1. Distorts the original distribution of the variable - right skewed above
2. If missingness is not important, it may mask the predictive power of the original variable by distorting its distribution.
3. If the number of NA is big, it will mask the true outlier of the distribution
4. If the number if NA is small, the replaced NA may be considered an outlier and pre-processed in a subsequent step fo feature engineering.
